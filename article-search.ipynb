# Which articles are missing images

```python
flist=glob.glob('articles/*.txt')
imglist=glob.glob('images/*.jpg')
```

```python
url_prefix='http://www.dinamalarnellai.com/site/news_folder/'
url_postfix='.jpg'
for fname in flist:
    article_id=fname.split('_')[2]
    with open(fname,encoding='utf-8') as f:
        text=f.read()
    matches=re.findall('news_folder/(.*?).jpg',text)
    img_matches=[i for i in imglist if article_id+'_' in i]
    if len(matches)!=len(img_matches):
        print(article_id, len(matches), len(img_matches))
        #for m in matches:
        #    print(m)
        #print()
        #for i in img_matches:
        #    print(i)
        #print()
        for m in matches:
            #print(m)
            found=False
            for i in img_matches:
                if m in i:
                    #print('Found!')
                    found=True
                    break
            if not found:
                print()
                print('Not found!')
                print(url_prefix+m+url_postfix)
```
